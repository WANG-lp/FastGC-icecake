```python
def showImageFromRaw(fname):
    with open(fname, 'r') as f:
        lines = f.readlines()
    height = int(lines[0])
    width = int(lines[1])
    print(height, width)
    raw = []
    for i in lines[2:]:
        sp = i.rstrip().split(" ")
        r,g,b = sp[0],sp[1],sp[2]
        raw.append([r,g,b])
    raw = np.asarray(raw, dtype=np.uint8)
    raw = raw.reshape(height, width, 3)
    print(raw.shape)
    return raw
# r = showImageFromRaw("/tmp/out2.bin")
r2 = showImageFromRaw("/tmp/out.bin")
imshow(r2)
```
